Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(11,6))
plt.plot(x, cos, label='Cosx', linestyle='solid', color='red')
plt.plot(x, cos3, label='3 Cosx', linestyle='dashed', color='green')
plt.plot(x, sin, label='Sin(x-90)', linestyle='dashdot', color='blue')
plt.plot(x, sin2, label='2 Sin x', linestyle='None', marker='x', color='#cece3d')

ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')

plt.legend(loc='center')

plt.show()
```
