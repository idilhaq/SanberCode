<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 
*   Series dapat menyimpan berbagai jenis tipe data, namun ketika digunakan, object type haru seragam. Berbeda dengan List yang dapat memuat berbagai jenis data type dan dapat memuat berbagai jenis data type saat digunakan. Begitu pula dengan Dictionary yang memiliki kemampuan untuk menyimpan berbagai jenis data type.
*   Dataframe dapat dikatakan sebagai kumpulan series yang tergabung membentuk tabel untuk kemudahan manipulasi data.

```python
# code here
import pandas as pd

x = pd.Series([9.0, 4.2, 5.5, 7.4])
print(x)
```
